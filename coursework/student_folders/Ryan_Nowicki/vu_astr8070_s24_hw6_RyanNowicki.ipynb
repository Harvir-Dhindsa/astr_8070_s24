```python
# Kludge to fix the bug with draw_ellipse in astroML
from matplotlib.patches import Ellipse

def draw_ellipse(mu, C, scales=[1, 2, 3], ax=None, **kwargs):
    if ax is None:
        ax = plt.gca()

    # find principal components and rotation angle of ellipse
    sigma_x2 = C[0, 0]
    sigma_y2 = C[1, 1]
    sigma_xy = C[0, 1]

    alpha = 0.5 * np.arctan2(2 * sigma_xy,
                          (sigma_x2 - sigma_y2))
    tmp1 = 0.5 * (sigma_x2 + sigma_y2)
    tmp2 = np.sqrt(0.25 * (sigma_x2 - sigma_y2) ** 2 + sigma_xy ** 2)

    sigma1 = np.sqrt(tmp1 + tmp2)
    sigma2 = np.sqrt(tmp1 - tmp2)

    for scale in scales:
        ax.add_patch(Ellipse(xy=(mu[0], mu[1]),
                          width=2 * scale * sigma1, height=2 * scale * sigma2,
                          angle=alpha * 180. / np.pi,
                          **kwargs))
```

```python
#Now fit a Gaussian mixture model. Compute the AIC and BIC for the number of GMM components between $1$ and $10$. 
#Plot these AIC and BIC values as a function of number of components. What is the optimal number according to the BIC? 
#Fit the data set with this optimal number of GMM components, and create a new plot that shows the data set, the GMM component mean locations, 
# and the $2$-sigma GMM component ellipses. *(Use the code below to help with plotting the ellipses.)*

#importing tools
from sklearn.mixture import GaussianMixture
```

```python
# Kludge to fix the bug with draw_ellipse in astroML
from matplotlib.patches import Ellipse

def draw_ellipse(mu, C, scales=[1, 2, 3], ax=None, **kwargs):
    if ax is None:
        ax = plt.gca()

    # find principal components and rotation angle of ellipse
    sigma_x2 = C[0, 0]
    sigma_y2 = C[1, 1]
    sigma_xy = C[0, 1]

    alpha = 0.5 * np.arctan2(2 * sigma_xy,
                          (sigma_x2 - sigma_y2))
    tmp1 = 0.5 * (sigma_x2 + sigma_y2)
    tmp2 = np.sqrt(0.25 * (sigma_x2 - sigma_y2) ** 2 + sigma_xy ** 2)

    sigma1 = np.sqrt(tmp1 + tmp2)
    sigma2 = np.sqrt(tmp1 - tmp2)

    for scale in scales:
        ax.add_patch(Ellipse(xy=(mu[0], mu[1]),
                          width=2 * scale * sigma1, height=2 * scale * sigma2,
                          angle=alpha * 180. / np.pi,
                          **kwargs))
```

```python
#5. Now try Kmeans clustering with $5$ clusters. 
 #   - Use the `StandardScalar` to scale the data.
 #   - Fit the Kmeans model.
 #   - Find the cluster centers, and then inverse transform thse cluster centers back to the original unscaled coordinates.
 #   - Predict the labels for the scaled data.
 #   - Finally, make a two-panel side by side plot showing (a) the original data colored by its labels, then (b) the data colored by the Kmeans predicted labels, with cluster centers superposed.
from sklearn.cluster import KMeans
from sklearn import preprocessing
```

```python
#fitting the Kmeans
clf = KMeans(n_clusters=5, n_init=10) #Using 5 clusters
clf.fit(data[:,:2]) #fitting the data
centers = clf.cluster_centers_ #location of the clusters
labels = clf.predict(data[:,:2]) #labels for each of the points
```

```python
#FROM LECTURE 14
# plot the data color-coded by cluster id
colors = ['m', 'r', 'y', 'g', 'b']
for ii in range(5):
    plt.scatter(data[:,:2][labels==ii,0], data[:,:2][labels==ii,1], 
                color=colors[ii])
plt.title('KMeans WITHOUT StandardScalar')
```

```python
#instantiate StandardScalar to scale the data properly
scaler = preprocessing.StandardScaler()
scaled_data = scaler.fit_transform(data[:,:2])
```

```python
#fitting the Kmeans and finding the centers
km = KMeans(n_clusters=5, n_init=10) #Using 5 clusters
km.fit(scaled_data) #fitting the data
```

```python
#find cluster centers and inverse transform back to unscaled coordinates
centers = km.cluster_centers_ #location of the clusters
unscaled_centers = scaler.inverse_transform(centers)
unscaled_data = scaler.inverse_transform(scaled_data)
```

```python
print(unscaled_centers)
```

```python
#predict the labels for the scaled data
labels = km.predict(scaled_data) #labels for each of the points
```

```python
#MAKING PLOT
colors = ['m', 'r', 'y', 'g', 'b']
datadata = data[:,:2]
datalabels = data[:,-1]

#set up figure
fig = plt.figure(figsize=(14, 4))

#subplot 1: the original data colored by its labels 
ax1 = fig.add_subplot(131)
ax1.set_xlim(-12,10)
ax1.set_ylim(-12,10)
for ii in range(5):#datalabels:
    #ii = int(ii)
    ax1.scatter(datadata[datalabels==ii,0], datadata[datalabels==ii,1], 
                color=colors[ii])
ax1.set_title('Data with Original Labels')

#subplot 2: data colored by the Kmeans predicted labels, with cluster centers superposed.
ax2 = fig.add_subplot(132)
ax2.set_xlim(-12,10)
ax2.set_ylim(-12,10)
for ii in range(5):#datalabels:
    #ii = int(ii)
    ax2.scatter(unscaled_data[labels==ii,0], unscaled_data[labels==ii,1], 
                color=colors[ii])
ax2.scatter(unscaled_centers[:,0], unscaled_centers[:,1], color='k', label='KMeans Cluster Centers')
ax2.legend()
ax2.set_title('KMeans Cluster Labels')
```

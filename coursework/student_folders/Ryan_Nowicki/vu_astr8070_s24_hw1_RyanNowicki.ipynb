## Problem 2



Consider a family with two children. Assume that the probability of females and males being born is equal, i.e. $p(F)=0.5$ and $p(M)=0.5$. Remember that "given" means your calculation is conditioned on some already fixed data or outcome. 

1) What is the probability of one male and one female **GIVEN** that the first child is male. 

2) What is the probability of two females **GIVEN** that at least one child is female.

3) What is the probability of two females **GIVEN** that the older child is female.

You will find it easiest to answer these questions by sketching a "tree diagram" to visualize the options for the oldest child, and the branching options for the youngest child. 

Either write your solution as Markdown/Latex below, or paste in a scanned image of your handwritten solution.

### Solution

1) *(To get you started)* If the first child is male, then the only unknown is the sex of the second child. We want the second child to be female, which has a probability of 0.5. Thus, $p(F,M |C_1=M) = 0.5$. 

2) In this scenario, there are four combinations of sibling order possible: F,F; F,M; M,F; and M,M. However, we are given that at least one child is female (I call this given $A$), which leaves us with the possble combinations of F,M; M,F; and F,F. Of these combinations, 1 in 3 have 2 females. Therefore, our probability that $p(F,F|A)=\frac{1}{3}$.

3) Since the probability that a child is either male or female is 0.5 and independent of the identity of their siblings (older and younger), when we are given that the older sibling is female, that leaves the probability of the younger child being female is 0.5. Expressed mathematically, $p(F,F |C_1=F) = 0.5$.

1) Use `numpy` to draw $1000$ random samples from a uniform distribution between $0.1$ and $10$, and store these samples as $x$.

2) Use matplotlib to make a histogram of these samples. Try to make this as aesthetically pleasing as possible (within reason).

3) Compute the base-10 log of your array $x$, and store this as $y$.

4) Make another histogram for $y$. Using the equation to transform probability distributions in `Lecture_2`, write here what the theoretical pdf of $y$ is, and overplot it onto your histogram.

5) Compute the mean of $x$ and the mean of $y$. Now compute the median of $x$ and the median of $y$. You should note that the means are different, but the medians (as it is a cumulative statistic) are the same. The mean is affected by the scale of the sample values, but the median only depends on the ordering of the samples. Monotonic transformations (like taking the log) do not change the ordering of samples.

### Solution

```python
#Part 1: numpy to get numbers
import numpy as np
x = np.random.uniform(0.1,10,1000) #args are low bound, high bound, and quantity
print(x)
```

```python
#Part 2: plotting as a histogram
import matplotlib.pyplot as plt

plt.hist(x, bins=10, color='teal')
plt.title('Random Numbers from Uniform Distribution from 0.1 to 10; 10 Bins')
plt.xlabel('Value of Random Number Generated from Uniform Distribution')
plt.ylabel('Samples Count')
```

```python
#Part 3:
y = np.log10(x)
print(y)
```

```python
#Part 4: plotting y as histogram
plt.hist(y, bins=10, color='turquoise')
plt.title('$Log_{10}$ Random Numbers from Uniform Distribution from 0.1 to 10; 10 Bins')
plt.xlabel('Log$_{10}$ Value of Random Number Generated from Uniform Distribution')
plt.ylabel('Samples Count')

#Using the equation to transform probability distributions in Lecture_2, write here what the theoretical pdf of is, and overplot it onto your histogram.
#p(x)dx = p(y)dy or p(y)=|dx/dy|*p(x)
px = 1/(10-0.1) #Uniform Distribution
#y = log10(x) so x=10^y and dx/dy = ln(10)*10^y = ln(10)*x
#therefore: py = np.abs(dxdy) * px

#plotting theoretical pdf (remember this log plot's horizontal axis is y value
horiz_val=[]
for i in range(-100,100):
    horiz_val.append(i/100)
horiz_val=np.array(horiz_val)
dxdy = (np.log(10))*(10**horiz_val)
yval    = 100 * np.abs(dxdy) * px #added scaling factor of 100 so curve is visible on plot given the number of samples
yval_unscaled = np.abs(dxdy) * px #without scaling factor
plt.plot(xval,yval, color='r')
plt.plot(xval,yval_unscaled, color='m')
```

```python
#Part 4: plotting y as histogram
plt.hist(y, bins=1000, color='turquoise')
plt.title('$Log_{10}$ Random Numbers from Uniform Distribution from 0.1 to 10; 1000 Bins')
plt.xlabel('Log$_{10}$ Value of Random Number Generated from Uniform Distribution')
plt.ylabel('Samples Count')

#Using the equation to transform probability distributions in Lecture_2, write here what the theoretical pdf of is, and overplot it onto your histogram.
#p(x)dx = p(y)dy or p(y)=|dx/dy|*p(x)
px = 1/(10-0.1) #Uniform Distribution
#y = log10(x) so x=10^y and dx/dy = ln(10)*10^y = ln(10)*x
#therefore: py = np.abs(dxdy) * px

#plotting theoretical pdf (remember this log plot's horizontal axis is y value
horiz_val=[]
for i in range(-100,100):
    horiz_val.append(i/100)
horiz_val=np.array(horiz_val)
dxdy = (np.log(10))*(10**horiz_val)
yval = np.abs(dxdy) * px #no scale factor present
plt.plot(xval,yval, color='m')
```

```python
#Part 5: comparing means and medians
xmean = np.mean(x)
ymean = np.mean(y)
xmed  = np.median(x)
ymed  = np.median(y)

print('Mean of X (unchanged, log10):', xmean, np.log10(xmean))
print('Mean of Y:', ymean)
print('Median of X (unchanged, log10):', xmed, np.log10(xmed))
print('Median of Y:', ymed)
```

```python
%matplotlib inline
%config InlineBackend.figure_format='retina' # very useful command for high-res images
# invoke pyplot in matplotlib, give it an alias
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 3*np.pi, 300)
y = np.sin(x)
y2 = np.sin(x**2)
y3 = np.cos(3*x)
plt.plot(x, y, color='red', label=r'$\sin(x)$')
plt.plot(x, y2, color='blue', label=r'$\sin(x^2)$')
plt.plot(x, y3,color='green', label=r'$\cos(3x)$')
plt.title('Some functions')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(which='major')
plt.legend(loc='lower left');
## optional extras
plt.minorticks_on()
plt.tick_params(which='both', direction='in', tick2On=True)
```

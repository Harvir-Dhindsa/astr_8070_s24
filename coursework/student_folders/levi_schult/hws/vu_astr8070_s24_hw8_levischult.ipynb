````python
data = np.load('../../../homeworks/hw_data/hw8_data_1.npy')
print(data.dtype)
````

````python
colnames = ['u', 'g', 'r', 'i', 'z', 'specClass', 'redshift', 'redshift_err']
````

````python
# LSS making arrays
uming = data['u'] - data['g']
gminr = data['g'] - data['r']
rmini = data['r'] - data['i']
iminz = data['i'] - data['z']

zarr = data['redshift']
sclass = data['specClass']
````

````python
print('num gal = ', np.sum(sclass=='GALAXY'))
print('num qso = ', np.sum(sclass=='QSO'))
````

# 2

````python
classlabels = [1 if s=='QSO' else 0 for s in sclass]

lab=['GAL', 'QSO']
colors=['tomato', 'lightblue']
plt.scatter(uming, gminr, c=classlabels, cmap=matplotlib.colors.ListedColormap(colors))
cb = plt.colorbar()
loc = np.arange(0,max(classlabels),max(classlabels)/float(len(colors)))
cb.set_ticks(loc)
cb.set_ticklabels(lab)
plt.xlabel('u-g')
plt.ylabel('g-r')
plt.xlim(-0.5, 2.5)
plt.ylim(-0.5, 1.5)
plt.show()
````

# 3

````python
y = classlabels
X = np.vstack([uming, gminr, rmini, iminz, zarr]).T

X_train, X_test, y_train, y_test= train_test_split(X, y, train_size=0.8)
````

# 4

````python
# LSS borrowed heavily from L19

for i in range(5):
    gnb = GaussianNB()
    gnb.fit(X_train[:, 0:i+1], y_train)
    y_prob = gnb.predict_proba(X_test[:, 0:i+1])[:, 1]
    fpr, tpr, thresh = roc_curve(y_test, y_prob)

    # add (0, 0) as first point
    fpr = np.concatenate([[0], fpr])
    tpr = np.concatenate([[0], tpr])

    plt.plot(fpr, tpr, label=f'num features ={i+1}')

plt.title('Gaussian Naive Bayes')
plt.xlim(0, 0.01)
plt.xlabel('FP rate')
plt.ylabel('TP rate')
plt.legend()
plt.show()
````

* redshift seems to make the classification quite easy!

# 5

````python
# LSS borrowed heavily from L19

for i in range(5):
    model = QuadraticDiscriminantAnalysis()
    model.fit(X_train[:, 0:i+1], y_train)
    y_prob = model.predict_proba(X_test[:, 0:i+1])[:, 1]
    fpr, tpr, thresh = roc_curve(y_test, y_prob)

    # add (0, 0) as first point
    fpr = np.concatenate([[0], fpr])
    tpr = np.concatenate([[0], tpr])

    plt.plot(fpr, tpr, label=f'num features ={i+1}')

plt.title('Quadratic Discriminant Analysis')
plt.xlim(0, 0.01)
plt.xlabel('FP rate')
plt.ylabel('TP rate')
plt.legend()
plt.show()
````

# 6

````python
# LSS borrowed heavily from L19

for i in range(1, 11):
    model = GMMBayes(n_components=i)
    model.fit(X_train[:, :5], y_train)
    y_prob = model.predict_proba(X_test[:, :5])[:, 1]
    fpr, tpr, thresh = roc_curve(y_test, y_prob)

    # add (0, 0) as first point
    fpr = np.concatenate([[0], fpr])
    tpr = np.concatenate([[0], tpr])

    plt.plot(fpr, tpr, label=f'num components ={i}')

plt.title('GMM Bayes')
plt.xlim(0, 0.01)
plt.xlabel('FP rate')
plt.ylabel('TP rate')
plt.legend()
plt.show()
````

* 7 or 9 components seems to do well! Should prefer lower number of components maps for model
simplicity

# 7

````python
# LSS borrowed heavily from L19

for i in np.arange(5, 100, 10):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(X_train[:, :5], y_train)
    y_prob = model.predict_proba(X_test[:, :5])[:, 1]
    fpr, tpr, thresh = roc_curve(y_test, y_prob)

    # add (0, 0) as first point
    fpr = np.concatenate([[0], fpr])
    tpr = np.concatenate([[0], tpr])

    plt.plot(fpr, tpr, label=f'num components ={i}')

plt.title('KNN classifier')
plt.xlim(0, 0.01)
plt.xlabel('FP rate')
plt.ylabel('TP rate')
plt.legend()
plt.show()
````

* all models with more than 25 components perform similarly it seems!
